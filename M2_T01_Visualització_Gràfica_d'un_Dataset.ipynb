```python
# Importo las librerías necesarias para los ejercicios:
import pandas
import numpy
from matplotlib import pyplot
import seaborn
```

```python
# Especifico que no se oculten columnas, para poder analizarlas:
pandas.set_option('display.max_columns', None)
```

## Exercici 1
Resumeix gràficament el data set **DelayedFlights.csv**.

Crea almenys una visualització per:
- Una variable categòrica (UniqueCarrier)

```python
# Almaceno el archivo csv que he exportado en el ejercicio anterior en un dataframe, y lo imprimo (está en "output"):
airlines_df = pandas.read_csv('./output/DelayedFlights_edited.csv')

airlines_df
```

```python
# Para la variable "UniqueCarrier" creo una gráfica con la cantidad de vuelos correspondientes a cada aerolínea:

pyplot.figure(figsize = (16, 8))
pyplot.title("Cantidad de Vuelos por Aerolínea")

ax = seaborn.countplot(x='UniqueCarrier', data=airlines_df)

# Esto hace que se muestren los valores de cada columna:
for container in ax.containers:
    ax.bar_label(container)
    
ax.grid(axis='y', color='gray', linestyle='dashdot', alpha=0.3)
pyplot.show()
```

- Una variable numèrica (ArrDelay)

```python
# Para la variable "ArrDelay" creo que lo más adecuado es un histograma, para comprobar su distribución:

airlines_df['ArrDelay'].hist()
pyplot.show()
```

- Una variable numèrica i una categòrica (ArrDelay i UniqueCarrier)

```python
# Para las variables 'ArrDelay' y 'UniqueCarrier' creo que lo más adecuado es un gráfico de dispersión:

pyplot.figure(figsize = (16, 6))
pyplot.title("Retrasos en la LLegada al Destino por Aerolínea")

ax = seaborn.stripplot(x=airlines_df['UniqueCarrier'], y=airlines_df['ArrDelay'], alpha=0.4)

ax.grid(axis='y', color='gray', linestyle='dashdot', alpha=0.3)
pyplot.show()
```

- Dues variables numèriques (ArrDelay i DepDelay)

```python
# Para las variables 'ArrDelay' y 'DepDelay' (que están estrechamente relacionadas) creo un gráfico de dispersión:

pyplot.figure(figsize = (10, 6))
pyplot.title("Relación entre Retrasos de Salida y Retrasos de Llegada")

ax = seaborn.scatterplot(x=airlines_df['ArrDelay'], y=airlines_df['DepDelay'])

ax.grid(axis='both', color='gray', linestyle='dashdot', alpha=0.3)
pyplot.show()
```

- Tres variables (ArrDelay, DepDelay i UniqueCarrier)

```python
# Integro la variable 'UniqueCarrier' en el gráfico de dispersión anterior, con el parámetro "hue":

pyplot.figure(figsize = (15, 6))
pyplot.title("Relación entre Retrasos de Salida y Retrasos de Llegada, especificando la Aerolínea")

ax = seaborn.scatterplot(x=airlines_df['ArrDelay'], y=airlines_df['DepDelay'], hue=airlines_df['UniqueCarrier'])

ax.legend(bbox_to_anchor=(1, 1))
ax.grid(axis='both', color='gray', linestyle='dashdot', alpha=0.3)
pyplot.show()
```

- Més de tres variables (ArrDelay, DepDelay, AirTime i UniqueCarrier)

```python
# La única forma que se me ocurre de integrar 4 variables en un gráfico es con un "paiplot",
#   y en principio su sintaxis me obliga a crear un dataframe solo con esas 4 variables:

reduced_df = airlines_df[['ArrDelay','DepDelay','AirTime','UniqueCarrier']]

reduced_df
```

```python
# Creo el "pairplot", con "kde" como valor del parámetro "diag_kind", que es adecuado para variables numéricas
#   (no sería necesario crear un nuevo dataframe si no utilizara el parámetro "hue", pero creo que es mejor utilizarlo):

seaborn.pairplot(reduced_df, hue='UniqueCarrier', diag_kind='kde')
pyplot.show()
```

## Exercici 3
Integra les visualitzacions gràfiques, en la tasca 2, del Mòdul 1.

```python
# Creo un gráfico para representar la cantidad de vuelos con retraso por aerolínea
#   (en base a la columna 'Delayed', que creé en el ejercicio "M1_T02"):

fig = pyplot.figure(figsize = (16, 8))
pyplot.title("Cantidad de Vuelos con Retraso por Aerolínea")

ax = airlines_df.groupby('UniqueCarrier').sum()['Delayed'].plot.bar()

for container in ax.containers:
    ax.bar_label(container)
    
ax.grid(axis='y', color='gray', linestyle='dashdot', alpha=0.3)
pyplot.xlabel('Aerolínea')
pyplot.ylabel('Vuelos con Retraso')
fig.autofmt_xdate(rotation=0, ha='center')

pyplot.show()
```

```python
# Creo un gráfico para representar el tiempo de retraso acumulado por aerolínea
#   (en base a la columna 'TotalDelay', que creé en el ejercicio "M1_T02"):

fig = pyplot.figure(figsize = (16, 8))
pyplot.title("Tiempo de Retraso Acumulado por Aerolínea")

ax = airlines_df.groupby('UniqueCarrier').sum()['TotalDelay'].plot.bar()

for container in ax.containers:
    ax.bar_label(container)
    
ax.grid(axis='y', color='gray', linestyle='dashdot', alpha=0.3)
pyplot.xlabel('Aerolínea')
pyplot.ylabel('Tiempo de Retraso Acumulado')
fig.autofmt_xdate(rotation=0, ha='center')

pyplot.show()
```

```python
# Creo un gráfico para representar la velocidad media por aerolínea
#   (en base a la columna 'Velocity', que creé en el ejercicio "M1_T02"):

fig = pyplot.figure(figsize = (16, 8))
pyplot.title("Velocidad Media por Aerolínea")

ax = airlines_df.groupby('UniqueCarrier').mean()['Velocity'].plot.bar()

for container in ax.containers:
    ax.bar_label(container)
    
ax.grid(axis='y', color='gray', linestyle='dashdot', alpha=0.3)
pyplot.xlabel('Aerolínea')
pyplot.ylabel('Velocidad Media')
fig.autofmt_xdate(rotation=0, ha='center')

pyplot.show()
```
